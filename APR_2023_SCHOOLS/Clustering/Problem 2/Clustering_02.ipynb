```python
###   Problem statement    : To cluster the different varieties of wheat based on their kernel properties
###
###   Model implimentation : K-means clustering
###
###   Dataset              : Wheat seeds
###

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

# Load the Seeds dataset
seeds_df = pd.read_csv('https://raw.githubusercontent.com/Deepsphere-AI/DSAI_WorkShop/main/APR_2023_SCHOOLS/Clustering/Problem%202/Dataset/seeds_train.csv')

# Cluster based on Length of the kernel and width of the kernel
x = seeds_df.iloc[:,[3,4]].values

wcss = []
for cluster in range(1,11):
    kmeans = KMeans(n_clusters = cluster, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
print("")
print("**********************************  OUTPUT  **********************************************")
plt.plot(range(1,11), wcss, 'o--')
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
print("******************************************************************************************")
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y = kmeans.fit_predict(x)

seeds_df['Cluster'] = kmeans.labels_

plt.figure(figsize=(10, 7))
plt.scatter(x[y == 0, 0], x[y == 0, 1], s = 20, c = 'red', label = 'Kama')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s = 20, c = 'blue', label = 'Rosa')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s = 20, c = 'green', label = 'Canadian')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = 'yellow', label = 'Centroids')
plt.title('Clusters of variety seeds')
plt.xlabel('Length of kernel')
plt.ylabel('Width of kernel')
plt.legend()
plt.show()

# predict
df_test= pd.read_csv("https://raw.githubusercontent.com/Deepsphere-AI/DSAI_WorkShop/main/APR_2023_SCHOOLS/Clustering/Problem%202/Dataset/seeds_test.csv")
df_test.head()

df_test=x = df_test.iloc[:,[3,4]]

df_values=df_test.values

find=df_values
result=[]
for i in find:
  z=kmeans.predict([i])
  if z==[0]:
    result.append("Kama")
  elif z==[1]:
    result.append('Rosa')
  elif z==[2]:
    result.append('Canadian')
df_test['Cluster']=result
print("**********************************   RESULT  *********************************************")
print(df_test)
print("******************************************************************************************")

## cluster based on area and perimeter
seeds_df = pd.read_csv('https://raw.githubusercontent.com/Deepsphere-AI/DSAI_WorkShop/main/APR_2023_SCHOOLS/Clustering/Problem%202/Dataset/seeds_train.csv')

x = seeds_df.iloc[:,[0,1]].values
wcss = []
for cluster in range(1,11):
    kmeans = KMeans(n_clusters = cluster, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
print("")
print("")
plt.plot(range(1,11), wcss, 'o--')
plt.title('Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
print("******************************************************************************************")

kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y = kmeans.fit_predict(x)

plt.figure(figsize=(10,6))
plt.scatter(x[y == 0, 0], x[y == 0, 1], s = 20, c = 'red', label = 'Kama')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s = 20, c = 'blue', label = 'Canadian')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s = 20, c = 'green', label = 'Rosa')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = 'yellow', label = 'Centroids')
plt.title('Clusters of variety seeds')
plt.xlabel('Area of kernel')
plt.ylabel('Perimeter of kernel')
plt.legend()
plt.show()
# predict
df_test= pd.read_csv("https://raw.githubusercontent.com/Deepsphere-AI/DSAI_WorkShop/main/APR_2023_SCHOOLS/Clustering/Problem%202/Dataset/seeds_test.csv")
df_test=x = df_test.iloc[:,[0,1]]
df_values=df_test.values

find=df_values
result=[]
for i in find:
  z=kmeans.predict([i])
  if z==[0]:
    result.append("Kama")
  elif z==[1]:
    result.append('Canadian')
  elif z==[2]:
    result.append('Rosa')

df_test['Cluster']=result
print("**********************************   RESULT  *********************************************")
print("")
print(df_test)
print("******************************************************************************************")
```
