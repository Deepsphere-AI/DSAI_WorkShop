```python
#Apartment rent prediction
#
#To build a linear regression model to predict the selling price of a new house based on its size and number of bedrooms.
#
## Model implemented : Linear Regression
#
#Importing the required libraries
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns

# Load the training dataset
data=pd.read_csv('https://raw.githubusercontent.com/Deepsphere-AI/ML_Algorithms/main/Data/Regression/Apartment_price_predict_Train.csv')

X = data[['Size (in sq ft)','Bedrooms']]
y = data[['Selling Price (in $1000s)']]

# # splitting X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Create a linear regression object
lr = LinearRegression()

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Calculate the metrics
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)

# Print the metrics
print(f"R-squared: {r2}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
# # Predict the selling price of a new house
# size = 1700
# bedrooms = 2
# predicted_price = model.predict([[size, bedrooms]])
# print("Predicted selling price of a {0} sq ft, {1} bedroom house: ${2:.2f}k".format(size, bedrooms, predicted_price[0][0]))

print('\n********************************************\n')
#Data visualization
X = 'Size (in sq ft)'
y = 'Selling Price (in $1000s)'

sns.set_style('whitegrid')
sns.lmplot(x = X, y = y, data = data)

```
